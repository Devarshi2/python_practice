```python
# Sample data
data = {
    'A': [1, 2, None, 4],
    'B': [None, 6, 7, 8],
    'C': [9, 10, 11, 12]
}
df = pd.DataFrame(data)

# Count non-null values in each column
column_counts = df.count()

print("Column Counts:")
print(column_counts)
```

```python
# Count non-null values in each row
row_counts = df.count(axis=1)

print("\nRow Counts:")
print(row_counts)
```
